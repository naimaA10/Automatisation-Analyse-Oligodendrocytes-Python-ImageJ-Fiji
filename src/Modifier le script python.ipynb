
## Modification du script classification
Nous venons de voir comment récupérer les informations sur un fichier au format csv. Avec la macro de classification des oligodendrocytes nous avons plusieurs fichiers au format csv. Il faut donner le chemin du répertoire contenant l'ensemble des fichiers csv et changer les antislashs en slashs comme vu précédemment.

```python
csv_dir = r"C:\Users\naima.ammiche\Desktop\Données - traitement images\Classification plaqueA".replace("\\","/")
```

Ensuite récupérer l'ensemble des fichiers csv de ce répertoire et les mettre dans la variable **csv_files** sous forme de liste. Nous parcourons ici la liste **csv_files** les éléments qu'elle contient avec la boucle *for* et la méthode *print*. 

```python
# Récupérer l'ensemble des fichiers csv produit
csv_regex = os.path.join(csv_dir.replace("\\", "/"), "*.csv") # Une chaîne de caractère des nom des fichiers à sélectionner
csv_files = glob.glob(csv_regex) # Liste de l'ensemble des fichiers CSV

# Afficher les élements de la liste
for csv in csv_files:
    print(csv)
```

Nous pouvons voir que cette liste contient les chemins de chaque fichier csv qui a été obtenu avec la macro de classification d'oligodendrocytes. Nous venons donc de récupérer le chemin de tous les fichiers au format csv dans la liste **csv_files**. Parcourir la liste **csv_files** avec une boucle *for* et pour chaque fichier récupérer le nom du puits comme expliqué précédemment. Pour chaque nom de puits récupérer l'ajouter à la liste all_puits avec la méthode *append* et afficher un message avec l'ensemble de puits.

```python
# Liste vide
all_puits = []

# Parcourt le répertoire et pour chaque fichier 
for csv in csv_files:
    # Parcourir csv et ajouter à la liste chaque puits ex [A01, A02, B01 ...]
    puits = csv[-7:-4] 
    if puits not in all_puits:
        all_puits.append(puits) 

# Afficher la liste contenant l'ensemble des puits à analyser
print(f"L'ensemble des puits {all_puits}")
```

Parcourir ensuite la liste **all_puits** représentant l'ensemble des puits donc A01 puis A02 ... et enfin H10. Pour chaque puits parcourut par exemple lorsqu'on parcourt A01  on parcourt la liste **csv_files** (la liste de tous les fichiers csv). 

Vérifier si le nom du puits ici parcourut de la liste **all_puits** exemple A01 et celui présent sur le nom du fichier csv parcourut de la liste **csv_files**  exemple img1_A01.csv sont idendiques avec la ligne de code **if name_puits == name:** dans ce cas essayer d'ouvrir ce fichier avec **try** s'il est vide un message apparaît notamment le message présent dans **except**.

Ensuite la ligne de code **if csv == csv_files[0]:** permet de vérifier si le fichier parcourut actuellement exemple **img1_A01** correspond au premier élément de la liste **csv_files** dans ce cas il produit un data frame **df** dans le cas contraire il l'ajoutait au data frame existant avec la ligne **df = pd.concat(objs = [cluster_skeleton,df], axis =0)**. Le fichier csv exemple *img1_A01.csv* est supprimée après avoir été ajouté au data frame. A la fin de la boucle donc lorsque tous les fichiers csv sont parcourus de la liste **csv_files**, le data frame **df** est enregistré au format csv dont le nom est par exemple **A01_save** pour le puits A01.

La ligne de code **print(f"Puits {name}\n{df.describe()}\n\n")** permet d'afficher le nom du puits suivi du data frame associé après avoir concaténé. La fonction describe() permet comme expliqué plus haut d'obtenir les statistiques du data frame. Les symboles **\\n** signifient le retour à la ligne lors de l'affichage pour avoir une meilleure visibilité du message.

Enfin la dernière ligne de code **df.drop(df.index[::], inplace=True)** permet de supprimer le data frame df après avoir était enregistrer au format csv pour produire un nouveau data frame pour un autre puits de la liste **all_puits**. *Exemple après avoir enregistré A01_save.csv le data df est supprimé et le même processus est reproduit pour le puits A04 et concaténé les fichiers csv de ce même puits et ce jusqu'au dernier puits H10*.

```python
# Parcourt la liste des puits A01, A02, B01... 
for name in all_puits:
    
    # Parcourt les fichiers csv obtenus par la macro de classification des oligodendrocytes
    for csv in csv_files:
        name_puits = csv[-7:-4]
        
        if name_puits == name:                
            # Essayer d'ouvrir le fichier CSV en data frame et affiché un message lorsque le fichier est vide 
            # (probablement causé lorsque des reflets présent sur l'image empêche l'analyse)
            try:  
                cluster_skeleton = pd.read_csv(csv) 
            except:
                print(f"Ce fichier est vide :  {csv}") 
    
            # Si c'est le prermier fichier CSV crée un data frame, sinon l'ajouter au dataframe existant
            if csv == csv_files[0]:
                df = cluster_skeleton
            else :
                df = pd.concat(objs = [cluster_skeleton,df], axis =0)
            os.remove(csv) 

    # Enregistrer le fichier data contenant les résultats de l'ensemble du puits au format CSV
    df.to_csv(os.path.join(csv_dir.replace("\\", "/"), name+"_save.csv"), index = False) # Sauvegarder les résultats dans un seul fichier csv
    df = pd.read_csv(csv_dir.replace("\\", "/")+"\\"+name+"_save.csv") # Ouvrir le fichier csv à analyser # Ouvrir un fichier CSV seulement
    print(f"Puits {name}\n{df.describe()}\n\n")
    
    # Supprimer le dataframe, vous pouvez retrouver le data frame le fichier csv sauvegarder
    df.drop(df.index[::], inplace=True) 
```

Avant de commencer l'analyse, initialiser les variables représentant ici les liste par exemple si on veut 3 classes pour les jonctions avec moins de 20 jonctions, entre 20 et 30 et supérieur à 30 alors il faut 3 listes vide vous pouvez leur donner le nom de variable que vous souhaitez.  La variable index_classification contiendra les puits analysés.  

```python
# Initialisation
inf20 = []
entre20_30 = []
sup30 = []

# Légende: les puits analysé
index_classification = []
```

Les fichiers sont concaténés nous avons ainsi un fichier csv par puits qui regroupe en lignes les oligodendrocytes du puits et en colonne les différentes caractéristiques. Le même processus que précédement est réalisé c'est-à-dire récupérer les fichiers au format csv mais cette for il ne s'agit pas des fichiers dont le nom est *img1_A01.csv, img2_A01.csv, ... img1_H10.csv, img100_H10 etc.* mais des fichiers obtenus avec la concaténation donc *A01_save.csv, A04_save.csv ... H10_save.csv*

Parcourir la liste **csv_files** qui contient les fichiers *A01_save.csv, A04_save.csv ... H10_save.csv* et récupérer le nom du puits pour chaque fichier et l'ajouter à la liste **index_classification**. Ensuite pour chaque fichier csv essayer de l'ouvrir en data frame avec la ligne de code **pd.read_csv(csv)** dans le cas contraire afficher un message pour prévenir que le fichier est vide

Ensuite à partir du commentaire *# Caractéristique* vous pouvez modifier mettre autant de clusters que nécessaire par exemple 3 clusters dans cet exemple et pour chaque cluster récupéré les variables souhaiter inférieur à 2 jonctione <20, entre 20 et 30 jonctions >=20 & <30 et supérieur à 30 jonctions >=30.

Calculer ensuite le pourcentage d'oligodendrocytes pour chaque cluster. Ainsi les variables **value** contiennent un pourcentage, *exemple ici value1 contient le pourcentage du cluster 1 et donc du nombre de jonctions inférieur à 20*. L'étape suivante consiste à ajouter ces pourcentages aux listes correspondantes donc ajouter *value* à la liste *inf20* avec la ligne de code **inf20.append(value1)**. Cette ligne de code prend la liste **inf20** et ajoute la valeur **value1** à la fin de la liste. 

Enfin un affichage permet de voir pour chaque puits le pourcentage pour chaque cluster. La partie affichage comporte les lignes de code : 
- **print(f"\nPuits {puits} : \nDimension totale du data frame : {df.shape} : Il y a {df.shape[0]} oligodendrocytes dans ce puits. Il y a {df.shape[1]} caractéristiques étudiées sur ces oligodendrocytes")** qui affiche le nom du puits et les dimensions du data frame en précisant le nombre d'oligodendrocytes (=nombre de ligne) et le nombre de caractéristiques (=nombre de colonne).
- **print(f"<20 : {value1:.2f} %\nentre 20 et 30 : {value2:.2f}%\>=30 : {value3:.2f}%")** qui affiche le pourcentage de chaque cluster

```python
# Récupérer l'ensemble des fichiers csv dans la liste csv_files
csv_regex = os.path.join(csv_dir.replace("\\", "/"), "*.csv") # Une chaîne de caractère contenant nom des fichiers à sélectionner
csv_files = glob.glob(csv_regex) # Liste de l'ensemble des fichiers CSV

# Parcourt les fichiers csv produit à l'étape précèdente
for csv in csv_files:
    
    # Nom du puits
    puits = csv[-12:-9] 
    # Ajouter le puits analyser
    index_classification.append(puits)
    
    # Ouvrir le fichier csv si celui ci est vide un message apparaît
    try :
        df = pd.read_csv(csv)
    except:
        print(f"Ce fichier est vide :  {csv}") 
                
    # Caractéristique
    cluster1 = df[df['# Junctions'] < 20]
    cluster2 = df[(df['# Junctions'] >= 20) & (df['# Junctions'] < 30)]
    cluster3 = df[df['# Junctions'] > 30]
    
    
    # % d'oligodendrocytes pour chaque classe
    oligo_total = df.shape[0] # Nombre total d'oligodendrocytes dans le puits
    value1 = (cluster1.shape[0]/oligo_total)*100 # avec moins de 20  jonctions
    value2 = (cluster2.shape[0]/oligo_total)*100 # entre 150 et 300 branches
    value3 = (cluster3.shape[0]/oligo_total)*100 # entre 300 et 450 branches

    
    # Ajouter ces pourcentage aux listes correspondantes
    inf20.append(value1)
    entre20_30.append(value2)
    sup30.append(value3)

    # AFFICHAGE
    #Dimension du data frame et pourcentage de chaque classe d'oligodendrocytes
    print(f"\nPuits {puits} : \nDimension totale du data frame : {df.shape} : Il y a {df.shape[0]} oligodendrocytes dans ce puits. Il y a {df.shape[1]} caractéristiques étudiées sur ces oligodendrocytes")
    print(f"Pourcentage d'oligodendrocytes par cluster pour le puits {name}") 
    print(f"<20 : {value1:.2f} %\nentre 20 et 30 : {value2:.2f}%\>=30 : {value3:.2f}%")
    
    # Affichage
    df.drop(df.index[::], inplace=True) # Supprimer le dataframe
```
